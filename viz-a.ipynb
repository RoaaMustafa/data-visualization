```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python

car_crash_df = sns.load_dataset("car_crashes")
car_crash_df.head()
x = car_crash_df[car_crash_df.alcohol >6]
x
alcohol_by_abbrev = x.groupby("abbrev", as_index=False).alcohol.mean()
alcohol_by_abbrev

sns.barplot(x="abbrev",y="alcohol",data=alcohol_by_abbrev)
```

##  from this graph above I can conclude which abbrev has  most alcohol percentage

##  Find Thee most common alcohol percentage between 4 and 6

```python
sns.relplot(x='alcohol', y='total', data=car_crash_df)
```

## summary
